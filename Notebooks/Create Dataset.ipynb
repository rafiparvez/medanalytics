```python
def mergeDFs(orig_df,df,x):
    temp = df.loc[:,['subject_id','hadm_id',x]]
    temp2 = temp.groupby( ["subject_id", "hadm_id"] ).mean()
    temp2=temp2.reset_index()
    
    df_merged= pd.merge(orig_df,temp2, on=['subject_id', 'hadm_id'])
    print(orig_df.shape)
    print(df_merged.shape)
    return df_merged
```

```python
df_final=None

sevs=['sofa', 'sapsii', 'apsiii', 'oasis', 'lods']
for ctr, df in enumerate([df_sofa, df_sapsii, df_apsiii, df_oasis, df_lods]):
    if (ctr>=1):
        temp5 = mergeDFs(df_final,df,sevs[ctr])
        print("working for "+sevs[ctr])
        df_final = temp5
    else:
        df_final = mergeDFs(df_cleaned_master,df,sevs[ctr])
        print("working for "+sevs[ctr])
```
